```python
import random

d = [0,5,27,6,13,28,100,45,10,79]

def pick_a_number_from_list(d):
    sm = 0   #find sum of numbers in d
    for i in (d):
        sm += i

    d1 = []    #normalised values
    for j in range(len(d)):
        d1.append(d[j]/sm)

     
    d2 = []  # cumulative sum 
    for k in range(len(d1)):
        d2.append(sum(d1[:k+1]))
     


    r = random.uniform(0.0, 1.0)
    print(r)

    number = 0
    for m in range(len(d2)):
        if r >= d2[m] and r <d2[m+1] :
            number = d[m+1]
    return number



    
def sampling_based_on_magnitued():
    d = [0,5,27,6,13,28,100,45,10,79]
    d.sort()
    result = dict()
    print(d, sum(d))
    for n in range(1, 100):
        number = pick_a_number_from_list(d)
        print(number)
        if number not in result:
            result[number] = 1
        else:
            result[number]+=1
            
    print(result)

sampling_based_on_magnitued()
```
